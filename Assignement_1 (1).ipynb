```python
def reverse(x: int) -> int:
        x = abs(x)
        reversed_int = 0
        while x != 0:
            remainder = x % 10
            x //= 10
            reversed_int = reversed_int + remainder
        print(reversed_int)
reverse(789)
```

## 3. Final Correct Code



```python
# Final working solution
def reverse(x: int) -> int:
        MAX_INT = 2**31 - 1
        MIN_INT = -2**31
        sign = 1
        if x < 0:
            sign = -1
        x = abs(x)
        reversed_int = 0
        while x != 0:
            remainder = x % 10
            x //= 10
            reversed_int = reversed_int * 10 + remainder

        return (reversed_int * sign) if MIN_INT <= reversed_int <= MAX_INT else 0

reverse(-7846)
```

## 3. Final Correct Code



```python
# Final working solution
def twoSum( nums: list[int], target: int) -> list[int]:
        required_nums = {}
        for i, num in enumerate(nums):
            second_number = target - num

            if num + second_number == target and second_number in required_nums:
                return [required_nums[second_number], i]

            required_nums[num] = i
twoSum([5,7,9,10,25], 19)
```

```python
def isPalindrome(s: str) -> bool:
        cleaned_string = ""
        for char in s:
            if 'a' <= char <= 'z':
                cleaned_string += char
            elif 'A' <= char <= 'Z':
                cleaned_string += char.lower()

        print(cleaned_string)
        print(cleaned_string[::-1])
        return cleaned_string == cleaned_string[::-1]
isPalindrome("once Upon A TiMe!")
```

## 3. Final Correct Code



```python
# Final working solution
def isPalindrome(s: str) -> bool:
        cleaned_string = ""
        for char in s:
            if char.isalpha():
                cleaned_string += char.lower()

        return cleaned_string == cleaned_string[::-1]
isPalindrome("racE caR!")
```
